```python
def plot(states):
    fig = plt.figure();
    ax = fig.gca(projection='3d')
    X = []
    Y = []
    Z = []
    for i in range(9):
        X.append(i)
        Z.append([])
        for j in range(9):
            Y.append(j)
            Z[-1].append(states[i][j].value)

    surf = ax.plot_surface(X, Y, np.array(Z), cmap=cm.coolwarm,
                           linewidth=0, antialiased=True)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
```
